```python
# testdata_exploration.ipynb

import os
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# Set the paths
test_correct_dir = '../data/test/correct'
test_incorrect_dir = '../data/test/incorrect'

# Function to display images
def display_images_from_folder(folder, num_images=5):
    images = os.listdir(folder)[:num_images]
    plt.figure(figsize=(15, 5))
    for i, img_name in enumerate(images):
        img_path = os.path.join(folder, img_name)
        img = load_img(img_path, target_size=(64, 64))
        plt.subplot(1, num_images, i + 1)
        plt.imshow(img)
        plt.axis('off')
    plt.show()

# Display some images from the correct and incorrect folders
print("Correct Images:")
display_images_from_folder(test_correct_dir)

print("Incorrect Images:")
display_images_from_folder(test_incorrect_dir)

# Check the number of images in each class
num_correct = len(os.listdir(test_correct_dir))
num_incorrect = len(os.listdir(test_incorrect_dir))

print(f"Number of correct images: {num_correct}")
print(f"Number of incorrect images: {num_incorrect}")
```
