```python
import sys
import os

# Add the absolute paths of the models and scripts folders to sys.path

import tensorflow as tf
current_dir = os.getcwd()

sys.path.append(os.path.abspath(os.path.join(current_dir,'..')))
 
from models.testadvance import create_advanced_dnn  # Import from models directory

from scripts.train_model import load_data  # Import from scripts directory

# Load testing data
test_dir = '../data/test'
test_data = load_data(test_dir)

# Choose model architecture
model = create_advanced_dnn(input_shape=(64, 64, 3))  # Adjust input shape as needed

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])


history = model.fit(test_data, epochs=10)

# Plot testing history
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['loss'], label='loss')
plt.xlabel('Epoch')
plt.ylabel('Value')
plt.legend()
plt.title('Testing History')
plt.show()
```
