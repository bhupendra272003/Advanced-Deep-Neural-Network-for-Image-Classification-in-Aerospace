```python
#Model_training
import sys
import os

# Add the absolute paths of the models and scripts folders to sys.path

import tensorflow as tf
current_dir = os.getcwd()

sys.path.append(os.path.abspath(os.path.join(current_dir,'..')))
 
from models.advanced_dnn import create_advanced_dnn  # Import from models directory

from scripts.train_model import load_data  # Import from scripts directory

# Load training data
train_dir = '../data/train'
train_data = load_data(train_dir)

# Choose model architecture
model = create_advanced_dnn(input_shape=(64, 64, 3))  # Adjust input shape as needed

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(train_data, epochs=10)

# Plot training history
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['loss'], label='loss')
plt.xlabel('Epoch')
plt.ylabel('Value')
plt.legend()
plt.title('Training History')
plt.show()
```
